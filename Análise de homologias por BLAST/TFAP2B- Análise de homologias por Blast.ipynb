# Análise de homologias por BLAST

```python
from Bio.Blast import NCBIXML 
from Bio.Blast import NCBIWWW 
from Bio import SeqIO
```

```python
#TFAP2B em fasta
f_TFAP2B = SeqIO.read(open("TFAP2B_prot.fasta"), format="fasta") 
print (len(f_TFAP2B.seq))
```

#### Procura de genes homólogos

```python
result_handle = NCBIWWW.qblast("blastp", "nr", f_TFAP2B.format("fasta"))
```

```python
file_save = open("my_blast_TFAP2B.xml", "w")
file_save.write(result_handle.read()) 
file_save.close() 
result_handle.close()
```

```python
result_handle = open("my_blast_TFAP2B.xml")
record = NCBIXML.read(result_handle)
```

```python
print("Parâmetros")
print("Database: " , record.database)
print("Matrix: ", record.matrix)
print("Gap penalties: ", record.gap_penalties)
```

```python
e_value_threshold=0.0000001
c=len(f_TFAP2B.seq)

print("Sequências homólogas:")
print("")
for alignment in record.alignments:
    for hsp in alignment.hsps:
        n=hsp.align_length
        i=(hsp.identities/n)
        q=(n/c)
        p=(hsp.positives/n)
        if hsp.expect < e_value_threshold and q>=0.9 and i>=0.9:
            print("Acesso: " , alignment.accession)
            print("Hit id: " , alignment.hit_id, "\n")
            print("E-value: ",hsp.expect)
            print("Comprimento: ",hsp.align_length)
            print( "Identidade: ", (i*100) ,"%")       
            print( "Query cover: ", (q*100) ,"%")
            print("Positivies (propriedade quimica semelhante):", (p*100), "%", "\n" )
            print("Definição: " , alignment.hit_def, "\n")
            print("------Alinhamento-----", "Query start: ",hsp.query_start, "Sbjct start: ",hsp.sbjct_start)
            print(hsp.query[0:125])
            print(hsp.match[0:125])
            print(hsp.sbjct[0:125])
            print("")
            print(hsp.query[125:250])
            print(hsp.match[125:250])
            print(hsp.sbjct[125:250])
            print("")
            print(hsp.query[250:375])
            print(hsp.match[250:375])
            print(hsp.sbjct[250:375])
            print("")
            print(hsp.query[375:481])
            print(hsp.match[375:481])
            print(hsp.sbjct[375:481])
            print("")
            print("")
            print("")
```
