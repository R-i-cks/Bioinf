```python
import os
import json

import time
from Bio import Entrez, Medline
```

```python
def extrair_info(search_term, email):
    Entrez.email = email
    handle = Entrez.esearch(db="pubmed", term=search_term )
    record=Entrez.read(handle)
    handle.close()
    id_list=record["IdList"]
    handle=Entrez.efetch(db="pubmed", id=id_list, rettype="medline", retmode="text")
    records=Medline.parse(handle)
    results=[]
    for record in records:
        result={"title": record.get("TI", ""), "author": record.get("AU", ""), "abstract": record.get("AB", "")}
        results.append(result)
    handle.close()
    return (results)
```

```python
def gravarBD(bd,fnome):
    f=open(fnome, "w", encoding='utf-8')
    json.dump(bd,f)
    return bd
```

# Para o PNPLA3

```python
result_PNPLA3=extrair_info("PNPLA3","...@gmail.com" )
gravarBD(result_PNPLA3, "informaÃ§ao PNPLA3")
```
